### In an effort to make list more manageable, I will pull up the last 120 rows of data.

```python
pitchdf = pd.read_csv("../data/Raw_Pitching_Data.csv", encoding= 'utf-8')
pitchdf.tail(10)
```

### I am going to drop columns that are not needed for my purpose.

```python
pitchdf = pitchdf.drop(['GIDP', 'SF', 'SH', 'R', 'GF','BFP','BK','HBP','WP','IBB','BAOpp', 'BB', 'SO','HR','ER','H','lgID','stint','SV'], axis=1)
```

```python
pitchdf.tail(2)
```

### I only want to include starting pitchers, so any pitcher's season where Games Started does not equal Games will be removed. 

```python
pitchdf = pitchdf[pitchdf.GS == pitchdf.G]
```

```python
pitchdf.tail(2)
```

### I am looking to see if there are any missing values left in the data

```python
print(pitchdf[pitchdf.isna().any(axis=1)])
```

### I will only include pitchers' seasons that recorded an out and therefore had an ERA. I am going to drop all rows with a NaN value

```python
pitchdf = pitchdf.dropna(axis = 'rows');
print(pitchdf[pitchdf.isna().any(axis=1)])
```

### My data has been wrangled.

```python
pitchdf.to_csv("../data/Clean_Pitching_Data.csv")
```
